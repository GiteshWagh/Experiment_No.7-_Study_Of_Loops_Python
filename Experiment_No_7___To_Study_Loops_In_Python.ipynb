```python
# Problem :
# Write a code to calculate the factorial of the number.

# Solution :
n =  int(input("Enter the number : "))

i = 1
fact = 1

while i<=n:
  fact = fact*i
  i=i+1

print(fact)
```

```python
# Problem :
# Write a code to calculate the Fibonacci Series using while loop.

# Solution :
n = int(input("Enter the number of itretions: "))

i = 1
a = 0
b=1

while i <= n:
  c = a+b
  print(a)
  a=b
  b=c
  i=i+1
```

```python
# Problem :
# Write a Python code to display that the given input is palidrome or not.

# Solution :
# Taking input from the user
String = input("Enter the input : ")

# Converting the string to lower case
String1 = String.lower()
EmptyStr=""
i = len(String1)-1
while i >= 0:
  String1_Index = (String1[i])
  EmptyStr = EmptyStr + String1_Index
  i=i-1

# Printing if the number is palidrome
if EmptyStr == String1:
    print("Yes. It is palidrome")

# Printing if the number is not palidrome
else:
    print("No. It is not a palidrome")
```

```python
# Problem :
# Write a Python program to search an element in a list.

# Solution :
# Method 1 : Using While loop
list = [1,2,3,4,5,6,7,8,9]
n = int(input("Enter the number to search : "))
while True:
  if n in list:
    print("Element found")
    break
  else:
    print("Element not found")

# Method 2: Using conditional statement
list = [1,2,3,4,5,6,7,8,9]
n = int(input("Enter the number to search : "))

if n in list:
  print("Element is present.")
else:
  print("Element is not present.")
```

```python
# Introduction to continue statement
# Use : When we want to skip a perticular iteration and continue next one.

# Problem :
# Write a Python program to Print numbers from 1 to 10 (skip 5)

# Solution :
Variable=1
while Variable<=10:
  if Variable==5:
    Variable = Variable + 1
    continue
  print(Variable,end=" ")
  Variable = Variable + 1
```

```python
# Problem :
# Write a Python programm to accept three digits and print all possible combination.

# Solution :
num1 = (input("Enter the first number : "))
num2 = (input("Enter the second number : "))
num3 = (input("Enter the third number : "))
string = num1+num2+num3

for i in range(3):
  for j in range(3):
    for k in range(3):
      if string[i] != string[j] and string[j] != string[k] and string[k] != string[i]:
        print(string[i],string[j],string[k])
```

```python
# Problem Statement :
# Write a code to find ANY Armstrong Numbers. Take input from the user for digits

# Solution :
digits = int(input("Find the armstrongs numbers of any digits : ")) # Accepting the input from the user

start = 10**(digits - 1)  # Calculating the starting point
end = 10**digits         # Calculating the ending point

for Number in range(start, end):
    strNum = str(Number)
    power = len(strNum)
    total = sum(int(i)**power for i in strNum)

    if total == Number:
        print(f"{Number} is an armstrong number") # Displaying is the number is armstrong
```

```python
# Problem Statement :
# Write a code to print all the prime numbers in given range.

# Solution :
n = int(input("Enter the starting number : "))
m = int(input("Enter the ending number : "))
for num in range(n,m+1):
  for i in range(2,num):
    if num%i==0:
      break
  else:
    print(num,end=" ")
```
